```python
import pandas as pd
df=pd.read_csv(r"C:\Users\unais\Desktop\nlpproject\nlpproject\redditpreprocess.csv")
df
```

```python
import gensim.downloader as api
import numpy as np
wv = api.load('glove-twitter-100')
```

```python

df['text'] = df['text'].apply(lambda x: np.mean([wv[token] for token in x if token in wv] or [np.zeros(wv.vector_size)], axis=0))
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['label_adr'] = label_encoder.fit_transform(df['label_adr'])
```

```python
import numpy as np
X = np.vstack(df['text'])  # Convert list of arrays to a matrix
y = df['label_adr']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.svm import SVC
#from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
rr=SVC()
#dt=MultinomialNB()
rd=DecisionTreeClassifier()
rf=RandomForestClassifier()
lst=[rr,rd,rf]
for i in lst:
  i.fit(X_train,y_train)
  y_pred1=i.predict(X_test)
  print('**************')
  print('**************')
  print(classification_report(y_test,y_pred1))
```
